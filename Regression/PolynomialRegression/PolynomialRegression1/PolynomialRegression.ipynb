<div class="alert alert-danger" role="alert">
ðŸ“¤ Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
```

<div class="alert alert-danger" role="alert"> 
ðŸ’¾ Read CSV file

```python
df = pd.read_csv('weatherData.csv')
df
```

<div class="alert alert-danger" role="alert"> 
ðŸ”Ž Seperate x and y

```python
x = df[['Pressure (millibars)']]
x
```

```python
y = df[['Humidity']]
y
```

* **the number of dataa is small so we dont seperate the train and test split**

<div class="alert alert-danger" role="alert"> 
ðŸ”Ž Polynomial Features

```python
poly_model = PolynomialFeatures(degree=2,include_bias=True)
poly_model
```

```python
poly_f = poly_model.fit_transform(x)
poly_f
```

```python
poly_f_df = pd.DataFrame(poly_f)
poly_f_df.columns = ['1','X','X^2']
poly_f_df
```

<div class="alert alert-danger" role="alert"> 
ðŸ”Ž LR Model

```python
LR = LinearRegression()
LR
```

```python
LR.fit(poly_f,y)
```

<div class="alert alert-danger" role="alert"> 
ðŸ”Ž Prediction

```python
y_pred = LR.predict(poly_f)
y_pred
```

```python
y
```

<div class="alert alert-danger" role="alert"> 
ðŸ”Ž Polynomial plot

```python
plt.scatter(x,y,color='purple')
plt.plot(x,y_pred,color='k')
plt.title('Polynomial',fontsize=20)
plt.xlabel('Pressure (millibars)',fontsize=13)
plt.ylabel('Humidity',fontsize=13)
plt.show()
```

<div class="alert alert-danger" role="alert"> 
ðŸ”Ž MAE

```python
metrics.mean_absolute_error(y,y_pred)
```

<div class="alert alert-danger" role="alert"> 
ðŸ”Ž Weight & Bias

```python
print(f"the theta is : {LR.coef_}")
print(f"the theta0 is : {LR.intercept_}")
```

* **We had three features, so we have three weights and a bias**
